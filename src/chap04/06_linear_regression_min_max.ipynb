```python
import tensorflow as tf
import numpy as np

# open, high, volume, low, close
xy = np.array([[828.659973, 833.450012, 908100, 828.349976, 831.659973],
               [823.02002, 828.070007, 1828100, 821.655029, 828.070007],
               [819.929993, 824.400024, 1438100, 818.97998, 824.159973],
               [816, 820.958984, 1008100, 815.48999, 819.23999],
               [819.359985, 823, 1188100, 818.469971, 818.97998],
               [819, 823, 1198100, 816, 820.450012],
               [811.700012, 815.25, 1098100, 809.780029, 813.669983],
               [809.51001, 816.659973, 1398100, 804.539978, 809.559998]])
```

```python
def min_max_scaler(data):
    numerator = data - np.min(data, axis = 0)
    denominator = np.max(data, axis = 0) - np.min(data, axis = 0)
    return numerator / (denominator + 1e-7) # denominator 의 값이 0이 될 수 있기 때문에 작은 값을 더해준다.
```

```python
xy_scaler = min_max_scaler(xy)
xy_scaler
```

```python
x_data = xy_scaler[:, :-1]
y_data = xy_scaler[:,[-1]]
```

```python
tf.model = tf.keras.Sequential()
tf.model.add(tf.keras.layers.Dense(units = 1, input_dim = x_data.shape[1]))
tf.model.add(tf.keras.layers.Activation('linear'))
```

```python
tf.model.compile(loss='mse', optimizer=tf.keras.optimizers.SGD(learning_rate=1e-5))
tf.model.summary()  # Param = features + b
```

```python
tf.model.fit(x_data, y_data, epochs = 100)
```
