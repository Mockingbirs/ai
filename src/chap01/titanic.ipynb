```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train = pd.read_csv("data/titanic/train.csv")
test = pd.read_csv("data/titanic/test.csv")
```

```python
train.shape[0]
```

```python
train.Survived.value_counts()
```

```python
sns.lmplot(x = 'Age', y = 'Survived', data = train, fit_reg = True , #축, y축, 데이터, line 없음
           scatter_kws = {"s":50}, #좌표 상의 점의 크기
           markers = ['o', "x"],
           hue = 'Sex') #예측값

#title
plt.title('Age and Sex in 2d plane')
plt.show()
```
